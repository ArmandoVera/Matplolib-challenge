```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "mouse_drug_data.csv"
clinical_trial_data_to_load = "clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_df = pd.read_csv(mouse_drug_data_to_load)
clinical_trial_df= pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
combined_data_df = pd.merge(mouse_drug_df, clinical_trial_df, how="outer", on="Mouse ID")
combined_data_df.head()

# Display the data table for preview

combined_data_df.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
groupy_1 = combined_data_df.groupby(["Drug", "Timepoint"]).mean()['Tumor Volume (mm3)']

# Convert to DataFrame
groupy_df = pd.DataFrame(groupy_1)

groupy_df.reset_index(inplace=True)

# Preview DataFrame
groupy_df.head()
```

```python
groupy_df.columns
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
groupstd = combined_data_df.groupby(["Drug", "Timepoint"]).sem()['Tumor Volume (mm3)']
#groupstd.head()
# Convert to DataFrame
groupstd_df = pd.DataFrame(groupstd)
groupstd_df.reset_index(inplace=True)
# Preview DataFrame
groupstd_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
pivoted_volmean_df = groupy_df.pivot(index='Timepoint', columns='Drug', values='Tumor Volume (mm3)')
# Preview that Reformatting worked
pivoted_volmean_df
```

```python
# Minor Data Munging to Re-Format the Data Frames
pivoted_volsem_df = groupstd_df.pivot(index='Timepoint', columns='Drug', values='Tumor Volume (mm3)')
# Preview that Reformatting worked
pivoted_volsem_df
```

```python
fig, ax = plt.subplots()
x_axis = pivoted_volmean_df.index.values
y_axis = pivoted_volmean_df.Capomulin.values
std_errors = groupstd_df[groupstd_df.Drug == 'Capomulin']['Tumor Volume (mm3)'].values
ax.errorbar(x_axis, y_axis, std_errors, fmt="o", label='Capomulin', color='red', linestyle='solid')
y_axis = pivoted_volmean_df.Infubinol.values
std_errors = groupstd_df[groupstd_df.Drug == 'Infubinol']['Tumor Volume (mm3)'].values
ax.errorbar(x_axis, y_axis, std_errors, fmt="^", label='Infubinol', color='blue', linestyle='solid')
y_axis = pivoted_volmean_df.Ketapril.values
std_errors = groupstd_df[groupstd_df.Drug == 'Ketapril']['Tumor Volume (mm3)'].values
ax.errorbar(x_axis, y_axis, std_errors, fmt="s", label='Ketapril', color='green', linestyle='solid')
y_axis = pivoted_volmean_df.Placebo.values
std_errors = groupstd_df[groupstd_df.Drug == 'Placebo']['Tumor Volume (mm3)'].values
ax.errorbar(x_axis, y_axis, std_errors, fmt="d", label='Placebo', color='black', linestyle='solid')
plt.title("Tumor Response to Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.legend(loc='topleft')
plt.grid()
```

```python
# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
group_MeanMet = combined_data_df.groupby(["Drug", "Timepoint"]).mean()['Metastatic Sites']
# Convert to DataFrame
group_MeanMet_df = pd.DataFrame(group_MeanMet)
group_MeanMet_df.reset_index(inplace=True)
# Preview DataFrame
group_MeanMet_df.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
group_SemnMet = combined_data_df.groupby(["Drug", "Timepoint"]).sem()['Metastatic Sites']
# Convert to DataFrame
group_SemMet_df = pd.DataFrame(group_SemnMet)
group_SemMet_df.reset_index(inplace=True)
# Preview DataFrame
group_SemMet_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
pivoted_Met_df = group_MeanMet_df.pivot(index='Timepoint', columns='Drug', values='Metastatic Sites')
# Preview that Reformatting worked
pivoted_Met_df
```

```python
# Generate the Plot (with Error Bars)
fig, ax = plt.subplots()
x_axis = pivoted_Met_df.index.values
y_axis = pivoted_Met_df.Capomulin.values
std_errors = group_SemMet_df[group_SemMet_df.Drug == 'Capomulin']['Metastatic Sites'].values
ax.errorbar(x_axis, y_axis, std_errors, fmt="o", label='Capomulin', color='red', linestyle='solid')
y_axis = pivoted_Met_df.Infubinol.values
std_errors = group_SemMet_df[group_SemMet_df.Drug == 'Infubinol']['Metastatic Sites'].values
ax.errorbar(x_axis, y_axis, std_errors, fmt="^", label='Infubinol', color='blue', linestyle='solid')
y_axis = pivoted_Met_df.Ketapril.values
std_errors = group_SemMet_df[group_SemMet_df.Drug == 'Ketapril']['Metastatic Sites'].values
ax.errorbar(x_axis, y_axis, std_errors, fmt="s", label='Ketapril', color='green', linestyle='solid')
y_axis = pivoted_Met_df.Placebo.values
std_errors = group_SemMet_df[group_SemMet_df.Drug == 'Placebo']['Metastatic Sites'].values
ax.errorbar(x_axis, y_axis, std_errors, fmt="d", label='Placebo', color='black', linestyle='solid')
plt.title("Mestastatic Spread During Treatment")
plt.xlabel("Treatment Duration Days")
plt.ylabel("Met Sites")
plt.legend(loc='topleft')
plt.grid()
# Save the Figure

# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
group_MouseCount = combined_data_df.groupby(["Drug", "Timepoint"]).count()['Mouse ID']
# Convert to DataFrame
group_MouseCount_df = pd.DataFrame(group_MouseCount)
group_MouseCount_df.reset_index(inplace=True)
# Preview DataFrame
group_MouseCount_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
pivoted_Mouse_df = group_MouseCount_df.pivot(index='Timepoint', columns='Drug', values= 'Mouse ID')
# Preview the Data Frame
pivoted_Mouse_df
```

```python
# Generate the Plot (Accounting for percentages)
drug_format_list = [('Capomulin','o','red'),('Infubinol','^','blue'),('Ketapril','s','green'),('Placebo','d','black')]
for drug,marker,colors in drug_format_list:
    total_mouse = pivoted_Mouse_df[drug][0]
    survival_rate = (pivoted_Mouse_df[drug]/total_mouse)*100
    survival_treatment_plt = plt.plot(pivoted_Mouse_df.index,survival_rate,
                                       marker=marker,ls='--',color=colors,linewidth=0.5)
    
plt.legend(loc='best')
plt.title('Survival During Treatment')
plt.xlabel('Times (Days)')
plt.ylabel('Survival Rate (%)')
plt.grid()
# Save the Figure
#plt.savefig(os.path.join('figures','survival_during_treatment.png'))
# Show the Figure
plt.show()
```
